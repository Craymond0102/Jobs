# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.datasets as datasets
import pandas as pd
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from sklearn.model_selection import train_test_split
```

# Load Dataset

```python
iris=datasets.load_iris()
#ONly taking two 
#X = iris.data[:, :2]

#TAKING all parameters
X=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
#dtree=DecisionTreeClassifier()
#dtree.fit(X,y)
```

# Classification decision tree

```python
dtree=DecisionTreeClassifier(criterion = 'entropy')
dtree.fit(X,y)
```

# Visualizing the Tree

```python
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

# Split dataset up
 - then use to test prediction accuracy

```python
test_percent = 0.3 # why am I using .3 here?

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_percent)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
for max_d in range(1,20,1):
    for min_samples in range(2,20,1):
        
        dtree=DecisionTreeClassifier(criterion ='entropy', max_depth=max_d, min_samples_split=min_samples)
        dtree.fit(X_train,y_train)

        prediction = dtree.predict(X_test)
        print("the max depth is :      " + str(max_d))
        print("The number of samples : " + str(min_samples))
        print("The prediction accuracy is: " + str(dtree.score(X_test,y_test)) +"%")
```

```python
#The best option for the decision tree when the tree is at a max depth of 3 and a number of samples of 3
#This gives a 97.7% accuracy and has the lowest amount desicions to make
```
